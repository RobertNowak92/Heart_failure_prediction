```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.linear_model
```

```python
heart_failure_data = pd.read_csv("data/heart_failure_clinical_records_dataset.csv")
```

```python
X = heart_failure_data.iloc[:, :-1]
X.head()
```

```python
y = heart_failure_data.iloc[:, -1]
y.head()
```

```python
X_ucz, X_test, y_ucz, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.35, random_state = 12345)
```

```python
y_ucz.mean()
```

```python
y_test.mean()
```

```python
def fit_regression(model, X_ucz, X_test, y_ucz, y_test):
    model.fit(X_ucz, y_ucz)
    y_ucz_pred = model.predict(X_ucz)
    y_test_pred = model.predict(X_test)
    r2 = sklearn.metrics.r2_score
    mse = sklearn.metrics.mean_squared_error
    mae = sklearn.metrics.mean_absolute_error
    return {
        "r_score_u": r2(y_ucz, y_ucz_pred),
        "r_score_t": r2(y_test, y_test_pred),
        "MSE_u": mse(y_ucz, y_ucz_pred),
        "MSE_t": mse(y_test, y_test_pred),
        "MAE_u": mae(y_ucz, y_ucz_pred),
        "MAE_t": mae(y_test, y_test_pred)
    }
```

```python
params = ["Reg. liniowa"]
res = [fit_regression(sklearn.linear_model.LinearRegression(),
                      X_ucz, X_test, y_ucz, y_test)]
df_results = pd.DataFrame(res, index=params)
```

```python
df_results
```

```python
import scipy as sc

z_scores = sc.stats.zscore(heart_failure_data)
z_scores
```

```python
abs_z_scores = np.abs(z_scores)
filtered_z_scores = (abs_z_scores < 3).all(axis=1)
```

```python
filtered_z_scores
```

```python
heart_failure_data_outl = heart_failure_data[filtered_z_scores]
heart_failure_data_outl.describe()
```

```python
heart_failure_data.describe()
```

```python
sns.pairplot(heart_failure_data_outl)
plt.show()
```

```python
corr_P_wout_outl = heart_failure_data_outl.corr("pearson")
corr_P_tri_wout_outl = corr_P_wout_outl.where(np.triu(np.ones(corr_P_wout_outl.shape, dtype=np.bool), k=1)).stack().sort_values()
corr_P_tri_wout_outl[np.abs(corr_P_tri_wout_outl)>0.4]
```

```python
X_outl = heart_failure_data_outl.iloc[:, :-1]
X_outl
```

```python
y_outl = heart_failure_data_outl.iloc[:, -1]
y_outl
```

```python
X_ucz_wo, X_test_wo, y_ucz_wo, y_test_wo = sklearn.model_selection.train_test_split(X_outl,
                                                                        y_outl,
                                                                        test_size=0.35,
                                                                        random_state=12345)
```

```python
y_ucz_wo.mean()
```

```python
y_test_wo.mean()
```

```python
params.append("Reg. liniowa wo")
res.append(fit_regression(sklearn.linear_model.LinearRegression(),
                          X_ucz_wo, X_test_wo, y_ucz_wo, y_test_wo))
df_results = pd.DataFrame(res, index=params)
df_results
```

```python
import sklearn.preprocessing
```

```python
X_ucz, X_test, y_ucz, y_test = sklearn.model_selection.train_test_split(X,
                                                                        y,
                                                                        test_size=0.35,
                                                                        random_state=12345)
```

```python
wielomian2 = sklearn.preprocessing.PolynomialFeatures(degree=2, include_bias=False)
X2_ucz = wielomian2.fit_transform(X_ucz)
X2_test = wielomian2.fit_transform(X_test)
```

```python
params.append("Reg. liniowa wielomianowa 2 stopnia")
res.append(fit_regression(sklearn.linear_model.LinearRegression(),
                          X2_ucz, X2_test, y_ucz, y_test))
df_results = pd.DataFrame(res, index=params)
df_results
```
