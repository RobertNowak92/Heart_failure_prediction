```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_predict
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from imblearn.over_sampling import SMOTE
```

```python
heart_failure_data = pd.read_csv("data/heart_failure_clinical_records_dataset.csv")
heart_failure_data = heart_failure_data.drop('time', 1)
heart_failure_data.describe()
```

```python
X = heart_failure_data.iloc[:, :-1]
m = X.mean()
s = X.std()
X_std = (X - m)/s
X_std.head()
```

```python
y = heart_failure_data.iloc[:, -1]
y.head()
```

```python
X_resampled, y_resampled = SMOTE().fit_resample(X, y)
```

```python
X_resampled.describe()
```

```python
y_resampled.describe()
```

```python
predicted = cross_val_predict(LogisticRegression(), X_resampled, y_resampled, cv=10)
res = {
        "ACC":  metrics.accuracy_score(y, predicted),
        "P":    metrics.precision_score(y, predicted),
        "R":    metrics.recall_score(y, predicted),
        "F1":   metrics.f1_score(y, predicted),
    }

params = ["LogRegression"]
df_results = pd.DataFrame(res, index=params)
df_results
```

```python
disp = sklearn.metrics.ConfusionMatrixDisplay(sklearn.metrics.confusion_matrix(y, predicted), display_labels=None)
disp.plot()
plt.show()
```

```python
fpr, tpr, threshold = metrics.roc_curve(y, predicted)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
AUC = sklearn.metrics.roc_auc_score(y, predicted)
GINI = 2* AUC - 1
print("AUC = ", AUC, " GINI = ", GINI)
```

```python
heart_failure_data_without_outl = pd.read_csv("data/heart_failure_data_without_outl.csv")
#heart_failure_data_without_outl = heart_failure_data_without_outl.drop('time', 1)
heart_failure_data_without_outl.describe()
```

```python
X_outl = heart_failure_data_without_outl.iloc[:, :-1]
m = X_outl.mean()
s = X_outl.std()
X_outl_std = (X_outl - m)/s
X_outl_std.head()
```

```python
y_outl = heart_failure_data_without_outl.iloc[:, -1]
y_outl.head()
```

```python
predicted = cross_val_predict(LogisticRegression(), X_outl_std, y_outl, cv=10)
res = pd.DataFrame({
        "ACC":  sklearn.metrics.accuracy_score(y_outl, predicted),
        "P":    sklearn.metrics.precision_score(y_outl, predicted),
        "R":    sklearn.metrics.recall_score(y_outl, predicted),
        "F1":   sklearn.metrics.f1_score(y_outl, predicted),
        }, index = ["LogRegression_without_outl"])
df_results = pd.concat([res, df_results])
df_results
```

```python
disp = sklearn.metrics.ConfusionMatrixDisplay(sklearn.metrics.confusion_matrix(y_outl, predicted), display_labels=None)
disp.plot()
plt.show()
```

```python
fpr, tpr, threshold = metrics.roc_curve(y_outl, predicted)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
AUC = sklearn.metrics.roc_auc_score(y_outl, predicted)
GINI = 2* AUC - 1
print("AUC = ", AUC, " GINI = ", GINI)
```
