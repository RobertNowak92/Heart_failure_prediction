# Modeling and results

Goal of this project is to create best model for predicting hearth failure. Models that were created are:
- Logistic regression
- k-Nearest Neighbors
- Random Forest
- Neural Network
All models will be created by using sklearn library.

The most important is to correctly predict hearth failure for people from risk group. Wrong diagnosis could mean their death.  Other way around if patient with low probability of heart attack were diagnosed incorrectly, to be in high risk group, nothing threatens his life.

To estimate performance of selected models, recall and F1 score will be calculated. Recall – it describes how many patients with high risk were diagnosed correctly. F1 score describes overall performance of the algorithm. It is dependent on correct diagnosis of patients with high risk factor and on all other misdiagnosed (type of I and II errors).

Parameters of the models will be decided by grid search with cross-validation on the train set to ensure the most optimal models to be trained.

Standard data as well as data without outliers will be used to check if removing the outliers had positive effect. Since data is highly imbalanced also oversampling method - smote - will be tested to try to overcome this problem.

Time variable has no sense in prediction of heart attack because it is follow-up time we cannot know before. Therefore it will be dropped from data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from imblearn.over_sampling import SMOTE
```

## Functions
The fit_classifier function allows to quicly fit given model, predict target viariable, create metrics and confusion matrix .

```python
def fit_classifier(alg, X_train, X_test, y_train, y_test):
      
    alg.fit(X_train, y_train)
    y_pred_train = alg.predict(X_train)
    y_pred_test = alg.predict(X_test)
    cm = metrics.confusion_matrix(y_test, y_pred_test, labels=None)
    met = {
        "ACC_train":  metrics.accuracy_score(y_train, y_pred_train),
        "ACC_test": metrics.accuracy_score(y_test, y_pred_test),
        "P_train":    metrics.precision_score(y_train, y_pred_train),
        "P_test":   metrics.precision_score(y_test, y_pred_test),
        "R_train":    metrics.recall_score(y_train, y_pred_train),
        "R_test":   metrics.recall_score(y_test, y_pred_test),
        "F1_train":   metrics.f1_score(y_train, y_pred_train),
        "F1_test":  metrics.f1_score(y_test, y_pred_test)
        }
    return met, cm
```

## Data load and standarization
### Original data

```python
heart_failure_data = pd.read_csv("data/heart_failure_clinical_records_dataset.csv")
heart_failure_data = heart_failure_data.drop(columns = 'time')
heart_failure_data.describe()
```

```python
#standarize data
X = heart_failure_data.iloc[:, :-1]
m = X.mean()
s = X.std()
X_std = (X - m)/s
X_std.head()
```

```python
#assign target variable
y = heart_failure_data.iloc[:, -1]
y.value_counts()
```

### Data without outliers

```python
heart_failure_data_without_outl = pd.read_csv("data/heart_failure_data_without_outl.csv")
heart_failure_data_without_outl = heart_failure_data_without_outl.drop(columns = 'time')
heart_failure_data_without_outl.describe()
```

```python
#standarize data
X_outl = heart_failure_data_without_outl.iloc[:, :-1]
m = X_outl.mean()
s = X_outl.std()
X_outl_std = (X_outl - m)/s
X_outl_std.head()
```

```python
#assign target variable
y_outl = heart_failure_data_without_outl.iloc[:, -1]
y_outl.value_counts()
```

## Create train and validation datasets

```python
X_std_train, X_std_val, y_std_train, y_std_val = train_test_split(X_std, y, test_size = 0.2, random_state = 53200)
X_outl_std_train, X_outl_std_val, y_outl_std_train, y_outl_std_val = train_test_split(X_outl_std, y_outl, test_size = 0.2, 
                                                                                      random_state = 53200)
for i in [X_std_train, X_std_val, y_std_train.value_counts(), y_std_val.value_counts(), 
          X_outl_std_train, X_outl_std_val, y_outl_std_train.value_counts(), y_outl_std_val.value_counts()]:
    display(i)
```

### Resample of the original data by SMOTE

```python
X_smote_train, y_smote_train = SMOTE().fit_resample(X_std_train, y_std_train)
X_smote_train
```

```python
y_smote_train.value_counts()
```

## Models
### Logistic regression

#### Original data

```python
#grid search and cross-validation
log_reg_param = {'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
                 'penalty': [None, 'elasticnet', 'l1', 'l2'],
                 'C': [0.001, 0.01, 0.1, 1, 10, 100]}

log_reg_gscv = GridSearchCV(estimator = LogisticRegression(l1_ratio = 0.5),  
                           param_grid = log_reg_param,
                           scoring = ['recall', 'f1'],
                           cv = 5,
                           refit = 'f1')

log_reg_gscv.fit(X_std_train, y_std_train)
```

```python
#Best parameters based on grid search and cross-validation
display(pd.DataFrame(log_reg_gscv.cv_results_)
        .loc[pd.DataFrame(log_reg_gscv.cv_results_)['rank_test_f1'] < 5]
        .sort_values('rank_test_f1')
       )
print(' Best parameters: ', log_reg_gscv.best_params_, '\n', 'Best f1 score: ', log_reg_gscv.best_score_)
```

```python
#training of the model and calculatin the metrics
params = ["Logistic regression"]
met, cm = fit_classifier(LogisticRegression(C = 0.1, penalty = 'l2', solver = 'liblinear'), 
                         X_std_train, X_std_val, y_std_train, y_std_val)
df_results = pd.DataFrame(met, index=params)
df_results
```

```python
#display the confusion matrix
disp = metrics.ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Logistic regression")
plt.show()
```

#### Data without outliers

```python
#grid search and cross-validation
log_reg_gscv.fit(X_outl_std_train, y_outl_std_train)
```

```python
#Best parameters based on grid search and cross-validation
display(pd.DataFrame(log_reg_gscv.cv_results_)
        .loc[pd.DataFrame(log_reg_gscv.cv_results_)['rank_test_f1'] < 5]
        .sort_values('rank_test_f1')
       )
print(' Best parameters: ', log_reg_gscv.best_params_,'\n','Best f1 score: ', log_reg_gscv.best_score_)
```

```python
#training of the model and calculatin the metrics
met, cm = fit_classifier(LogisticRegression(C = 0.001, penalty = 'l2', solver = 'liblinear'), 
                         X_outl_std_train, X_outl_std_val, y_outl_std_train, y_outl_std_val)
res = pd.DataFrame(met, index = ["Logistic regression without outliers"])
df_results = pd.concat([res, df_results])
df_results
```

```python
#display the confusion matrix
disp = metrics.ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Logistic regression without outliers")
plt.show()
```

### Neural network
To avoid overfitting number of hidden_layer_sizes must be low, it was decided empirically to be in range (2, 6) and also early_stopping is set to True.
#### Original data

```python
#grid search and cross-validation
NNet_params = {'solver': ['lbfgs', 'sgd', 'adam'], 
               'max_iter': [x for x in range(5000, 6100, 200)], 
               'alpha': 10.0 ** -np.arange(1, 5), 
               'hidden_layer_sizes':np.arange(2, 6)}

NNet_gscv = GridSearchCV(estimator = MLPClassifier(early_stopping = True),  
                           param_grid = NNet_params,
                           scoring = ['recall', 'f1'],
                           cv = 5,
                           refit = 'f1')

NNet_gscv.fit(X_std_train, y_std_train)
```

```python
#Best parameters based on grid search and cross-validation
display(pd.DataFrame(NNet_gscv.cv_results_)
        .loc[pd.DataFrame(NNet_gscv.cv_results_)['rank_test_f1'] < 5]
        .sort_values('rank_test_f1')
       )
print(' Best parameters: ', NNet_gscv.best_params_, '\n', 'Best f1 score: ', NNet_gscv.best_score_)
```

```python
#training of the model and calculatin the metrics
met, cm = fit_classifier(MLPClassifier(hidden_layer_sizes=(2,), alpha = 0.0001, max_iter = 5800, solver = 'lbfgs', 
                                        early_stopping = True), 
                         X_std_train, X_std_val, y_std_train, y_std_val)
res = pd.DataFrame(met, index = ["NNet"])
df_results = pd.concat([res, df_results])
df_results
```

```python
#display the confusion matrix
disp = metrics.ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("NNet")
plt.show()
```
